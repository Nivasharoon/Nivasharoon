```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
desc_stats = df_merged.describe()
```

```python
numerical_cols = ['overall', 'attack', 'midfield', 'defence', 'starting_xi_average_age', 'whole_team_average_age']

plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df_merged[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
# Select only numerical columns for correlation
numerical_cols = df_merged.select_dtypes(include=['float64', 'int64']).columns

# Generate the correlation matrix only for numerical columns
plt.figure(figsize=(10, 6))
corr_matrix = df_merged[numerical_cols].corr()

# Visualize the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='league_name', y='overall', data=df_merged)
plt.xticks(rotation=90)
plt.title('Overall Rating by League')
plt.show()
```
